```python
## SCRAPING

import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL yang akan di-scrape
url = 'https://top-1000-sekolah.ltmpt.ac.id/?Peringkat2022Search%5Brank_nas%5D=&Peringkat2022Search%5Bket%5D=&Peringkat2022Search%5Bnpsn%5D=&Peringkat2022Search%5Bnama_slta%5D=&Peringkat2022Search%5Bnilai_akhir%5D=&Peringkat2022Search%5Bprovinsi%5D=&Peringkat2022Search%5Bkab_kota%5D=Yogya&Peringkat2022Search%5Bjanis_slta%5D=&page=1&per-page=100'

# Mengunduh halaman HTML
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Mencari tabel dengan class "table-striped"
table = soup.find('table', class_='table table-striped table-bordered')

# Mengekstrak header tabel
header = [th.text.strip() for th in table.find_all('th')]

# Mengekstrak data dari setiap baris tabel
data = []
for row in table.find_all('tr'):
    row_data = [td.text.strip().replace('More..', '') for td in row.find_all('td')]
    if row_data:
        data.append(row_data)

# Membuat DataFrame dari data
df = pd.DataFrame(data[1:], columns=header)

# Menampilkan DataFrame
print(df)

df.to_csv('/content/drive/My Drive/datawebscrap/peringkat_sekolah_yogya_2022.csv', index=False)
```

```python
import pandas as pd

# Memuat data dari file CSV ke DataFrame
df = pd.read_csv('/content/drive/My Drive/datawebscrap/peringkat_sekolah_yogya_2022.csv')

# Menghapus kolom yang tidak diperlukan
df_cleaned = df[['Sekolah', 'Nilai Total']].copy()

# Menampilkan DataFrame yang sudah dibersihkan
print(df_cleaned)
```

```python
import matplotlib.pyplot as plt

# Bar Plot
plt.figure(figsize=(10, 6))
plt.bar(df_cleaned['Sekolah'], df_cleaned['Nilai Total'], color='skyblue')
plt.xlabel('Sekolah')
plt.ylabel('Nilai Total')
plt.title('Peringkat Sekolah di Kota Yogyakarta Tahun 2022')
plt.xticks(rotation=90) # Memutar label x agar lebih mudah dibaca
plt.tight_layout()
plt.show()

# Histogram
plt.figure(figsize=(8, 6))
plt.hist(df_cleaned['Nilai Total'], bins=20, color='lightgreen', edgecolor='black')
plt.xlabel('Nilai Total')
plt.ylabel('Frekuensi')
plt.title('Distribusi Nilai Total Sekolah di Kota Yogyakarta Tahun 2022')
plt.grid(True)
plt.tight_layout()
plt.show()

# Scatter Plot
plt.figure(figsize=(8, 6))
plt.scatter(df_cleaned['Sekolah'], df_cleaned['Nilai Total'], color='orange')
plt.xlabel('Sekolah')
plt.ylabel('Nilai Total')
plt.title('Hubungan antara Sekolah dan Nilai Total di Kota Yogyakarta Tahun 2022')
plt.xticks(rotation=90) # Memutar label x agar lebih mudah dibaca
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Membaca data dari file CSV
df = pd.read_csv('/content/drive/My Drive/datawebscrap/peringkat_sekolah_yogya_2022.csv')

# Membersihkan kolom 'Nilai Total'
df_cleaned = df[['Sekolah', 'Nilai Total']].copy()
df_cleaned['Nilai Total'] = df_cleaned['Nilai Total'].str.replace(',', '').astype(float)

# Menghitung rata-rata
mean_value = df_cleaned['Nilai Total'].mean()

# Plot garis
plt.figure(figsize=(14, 8))
plt.plot(df_cleaned['Sekolah'], df_cleaned['Nilai Total'], marker='o', color='blue', label='Nilai Total')
plt.axhline(mean_value, color='red', linestyle='--', label='Rata-rata')

# Anotasi untuk nilai rata-rata
plt.text(len(df_cleaned) / 2, mean_value + 5, f'Rata-rata: {mean_value:.2f}', color='red', fontsize=12, ha='center')

# Label dan judul
plt.xlabel('Sekolah')
plt.ylabel('Nilai Total')
plt.title('Peringkat Sekolah di Kota Yogyakarta Tahun 2022')
plt.xticks(rotation=45, ha='right')  # Memutar label x agar lebih mudah dibaca
plt.legend()
plt.tight_layout()
plt.show()
```
